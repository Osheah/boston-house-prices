## Background

How are housing prices determined? Why is it the average house price in Dublin €432k compared €115k in Longford (An Post and Ordnance Survey Ireland, 2019). There are socio environmental factors that make one property more desirable (hence more expensive) than a comparable property in a different area. Influencing factors could related to the property such as the age of the house, its BER rating or the number of rooms; Or the factors may be related to the local area such as the proximity to work/job opportunity or proximity to good school. The hedonic pricing model aims to capture and quantify these externalities that influence house price. 

The Boston housing dataset was created to investigate factors influencing house price, in particular levels of pollution. It made its first appearance in a paper by Harrison and Rubinfeld (1978) titled *'Hedonic prices and the demand for clean air'* (Harrison & Rubinfeld, 1978). It consisted of 506 observations across 14 measures from regions around Boston city USA. They developed a housing hedonic price index estimator to gauge the marginal maximum price people would pay to have clean air (Belsley, Kuh, & Welsch, 2004). Harrison and Rubinfeld (1978), collated 13 suspected influencing factors on the 14th measure,price. The factors they considered relevant to house price could be broadly classified as the dependent factor, price, structural factors, neighbourhood factors, accessibility factors and air pollution factors. 

The dependent factor MEDV measures the median value of owner occupied houses in the Boston area. It was taken from 1970 census traits. 

The two structural factors are RM and AGE. 
1. RM denotes the average number of rooms per owner occupier property. Harrison and Rubinfeld 1978 predicted more rooms would command higher prices i.e. a positive effect. Room number information came from the 1970 US census. 
2. AGE refers to house age. It measures proportion of houses build before 1940. Harrison and Rubinfeld reasoned that house age is indicative of structural quality. Older houses being less structurally sound than newer builds. A high level of old houses in an area was expected to reduce price, i.e. a negative relationship was initially predicted. 

There were eight neighbourhood factors, B, LSTAT, CRIM, ZN, INDUS, TAX, PTRATIO and CHAS. 
1. B measures the estimated proportion of black people in the area. Belsley,Kuh and Welsch (2004) noted that this was formulated by $(b-0.63)^2$ where b is the proportion of black people in the local population. Harrison and Rubinfeld suggested a parabolic relationship here. Initially the level of black people in a neighbourhood would depress the market as white people would leave the area but as demand increased for housing this would negate negative effects. This is reflects the period in which the report was written. The 1960 and 1970s was a period of civil unrest in the USA with a great deal of racism which echoes down to this day. How could you possibly measure the degree of blackness or whiteness a person contains. You simply can't because there is only one race and that is the human one.
2. LSTAT was a measure of the proportion of the population deemed 'lower status'. Low status was measured by 0.5 times the proportion of people lacking high school experience plus the proportion of men whose employment was classified as 'labourer'. Harrison and Rubinfeld, 1978 suggested this would be negatively correlated and have a greater effect at higher price areas than in lower price areas reflecting its logarithmic nature. This    
3. CRIM represents the crime rate per area taken from 1970 FBI figures. Harrison and Rubinfeld speculated (reasonably) that higher crime rates would depress house price. 
4. ZN represents the proportion of residential land zoned for plots greater than 25,000 square feet. This data was taken from 1972 Metropolitan Area planning commission report. Larger zoning areas limit the amount of smaller homes that can be constructed in the area. Harrison and Rubinfeld, 1978 predicted that larger zoning areas would show a positive relationship with house price. This is due to larger homes being more expensive and greater access to high zone level amenities such as sport fields which add to the area's desirability. 
5. INDUS measures the proportion of non retail business acres per town. These figures were taken from Vogt, Ivers  and Associates' Comprehensive land use inventory report by the department of commerce and development 1965. This variable serves as a way of representing the externalities associated with industrialised areas such as increased traffic, ugly buildings, noise levels etc. Harrison and Rubinsfeld expected this variable to negatively impact house prices. 
6. TAX refers to the full value tax rate on the property per 10,000. This variable was used to measure local public services costs. This measure was corrected by replacing nominal tax rates with local assessment ratios to reflect the full tax rate value in each area. Harrison and Rubinfeld expected this relationship to be negatively correlated with price. The data was taken from the 1970 Massachusetts Tax payers foundation. 
7. PTRATIO refers to the pupil teacher ratio per areas school district. This factor is used to measure public sector benefits. Harrison and Rubinfeld noted an unclear link between school quality and pupil teacher ratio. They suggested a low ratio should be indicative of the teachers attentiveness to each pupil. They expected a negative correlation between PTRATIO and price. The data was taken from the Massachusetts Department of Eduction 1971-72. 
6. The CHAS dummy variable is a binary marker that denotes whether or not the property tracts boarders the Charles River. If the property boarders the river it is recorded as a 1, otherwise 0. This data came from the 1970 US Census Tract maps.

The Boston Dataset contains two Accessibility measures DIS and RAD. 
1. DIS refers to the weighted distances from the property to 5 Boston area employment centres. Harrison and Rubinfeld, 1978 noted that economic urban land rent slopes predicted higher rent prices when properties were closer to high regions of employment. This data is recorded in logarithmic form and was taken from A. Schnare's 1973 phd dissertation 'An empirical analysis of the dimensions of neighbourhood quality'. Harrison and Rubinfeld expected this to be a negative log relationship. 
2. RAD denotes an index of accessibility to radial highways. A radial highway is one that directly connects to an urban area, in this case, Boston. The reason this measure was included was that access to quality roads prevents negative pollution measures swamping employment opportunities benefits gained by owning properties within commuting distances to work. Good arterial highways connecting urban centres was expected to be positively linked to house price. Harrison and Rubinfeld, 1978 expected this to be a positive log relationship. The data came from MIT's Boston Project.

There were two air pollution factors in the original paper, NOX and PART. However only NOX is included in the boston dataset probably as PART was not used in Harrison and Rubinfeld's pricing equation.
1. NOX is included in the Boston Dataset. NOX measures the concentration of Nitrogen Oxide in annual average concentration in parts per hundred million. The data was taken from Transportation and Air Shed Simulation Model (TASSIM)i.e simulated data. It would be expected that higher pollution levels would depress house prices however it could also reflect demand for housing in urban areas therefore showing a reverse effect.

### The pricing model
Harrison and Rubinfeld 1978 used the following pricing equation

$$log(MEDV) = a_{l} + a_{2}RM^2 + a_{3}AGE + a_{4}log(DIS) + a_{5}log(RAD) + a_{6}TAX + a_{7}PTRATI0 + a_{8}(B - 0.63)^2 + a_{9}log(LSTAT) + a_{10}CRIM + a_{ll}ZN + a_{l2}INDUS + a_{13}CHAS + a_{14}NOX + \epsilon$$

rewriting this in the order it appears in the boston dataset gives the following.

$$log(MEDV) = a_{l} + a_{2}CRIM + a_{3}ZN + a_{4}INDUS + a_{5}CHAS + a_{6}NOX + a_{7}RM^2 + a_{8}AGE + a_{9}log(DIS) + a_{10}log(RAD) + a_{ll}TAX + a_{l2}PTRATIO + a_{13}(B-0.63)^2 + a_{14}log(LSTAT) + \epsilon$$

H&auml;dle and Simer 2007 suggested the following model

$$log(MEDV) = a_{l} + a_{2}log(CRIM) + a_{3}\frac{ZN}{10} + a_{4}log(INDUS) + a_{5}CHAS + a_{6}log(NOX) + a_{7}log(RM) + a_{8}(\frac{AGE^{2.5}}{10000}) + a_{9}log(DIS) + a_{10}log(RAD) + a_{ll}log(TAX) + a_{l2}\frac{e^{0.4\times PTRATIO}}{1000} + a_{13}\frac{B}{100} + a_{14}\sqrt{LSTAT} + \epsilon$$

```python
def hide_current_axis(*args, **kwds):
    plt.gca().set_visible(False)

g = sns.pairplot(boston)
g.map_upper(hide_current_axis)
g.map_diag(hide_current_axis)
plt.show()
# taken from  https://stackoverflow.com/questions/34087126/plot-lower-triangle-in-a-seaborn-pairgrid
```

This gives a broad overview of the patterns between the various variables. The bottom line plots MEDV as y against the other variables as x. Next, plot the scatter plots with correlation and pick out interesting relationships.


```python
# Function to calculate correlation coefficient between two arrays
# https://towardsdatascience.com/visualizing-data-with-pair-plots-in-python-f228cf529166
def corr(x, y, **kwargs):
    
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))
    
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.2, 0.95), size = 20, xycoords = ax.transAxes)
    
# Create a pair grid instance
grid = sns.PairGrid(data= boston,
                    vars = boston.columns, height = 3)

# Map the plots to the locations
grid = grid.map_lower(plt.scatter, color = 'darkred')

grid = grid.map_lower(corr)
grid = grid.map_upper(hide_current_axis)
grid = grid.map_diag(hide_current_axis);
```

A strong correlation is r greater than +/- 0.7 and a moderate correlation is +/- 0.5 so assume anything over +/-0.5 is of interest to the pricing model.
